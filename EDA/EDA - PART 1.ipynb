```python
#missing values
df.isna().sum()
```

```python
features_with_na = [features for features in df.columns if df[features].isna().sum()>=1]
```

```python
#these are the features with NAN values.
features_with_na
```

```python
# df['Electrical'].isnull().mean()
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean()*100, 4), '% missing values')
```

```python
sns.boxplot(df['SalePrice'])
```

```python
#let's  plot some amazing diagram
data = df.copy()
for feature in features_with_na:
    #let's make a variable that indicate 1 if the observation was missing or 0 otherwise
    data[feature] = np.where(data[feature].isnull(), 1,0)
    
    #lets calculate the mean sales price where information is missing
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and the dependent variable is clearly visible.So we need to replace these NAN values with something meaningful which we will do in feature Engineering section.

### Numrical Variables

```python
#list of numerical variable
df['SaleType'].dtype !="object"
```

```python
numerical_features = [feature for feature in df.columns if df[feature].dtypes!='object']
```

```python
print(len(numerical_features))
df[numerical_features].head()
```

```python
df.info()
```

```python
# Temporal Variable(Eg: Datetime Variables)
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature, df[feature].unique)
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
#Here we will compare the difference between all year feature with SalesPrices
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')   
        plt.show()
```

#### Observation:
- Within all the year information, it si clearly visible that the New Houses are more costlier that the older one. looks like a paret or power law distribution.

```python
#numerical_variables - 2Types
#1. Continuous variable and Discrete variable

discrete_feature = [feature for feature in numerical_features if len(df[feature].unique() <=25)]
len(discrete_feature)
```

```python
df[discrete_feature].head()
```

```python
#find the relationship between Discrete and Sales Price
data = df.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
```

There is a relationship between Discrete Feature adn Sale Price

```python
## continous Varibales
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature]
```

```python
len(continous_feature)
```

```python
#let's analyze the continous values by creating histogram to understand the distribution
data = df.copy()
for feature in continous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

### EDA Part 2


```python
# we will be using Logarithmic transformation
data = df.copy()
for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature], data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
## Outliers

for feeature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel = (feature)
        plt.title =(feature)
        plt.show()
```
