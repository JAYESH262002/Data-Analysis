Two ways for the sublpots()

```python
#option 1 for plotting subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(8,6))
ax1.scatter(x, np.sin(x))
ax2.plot(x, x**2)
ax3.bar(np.linspace(1,6,10), np.random.rand(10))
ax4.hist(np.random.rand(20));
```

```python
#option 2 for plotting subplots
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(8,6))
ax[0, 0].plot(x, x**2)
ax[0, 1].scatter(np.random.rand(20), np.random.rand(20))
ax[1,0].hist(np.linspace(5, 20, 17))
ax[1,1].bar([1,2,5,6,9], [.2, .5, .7, .9, 1]);
```

# which should we use (pyplot vs myplotlib OO)
- ###  when plotting sometyhing quickly, use pyplot
- ###  when ploting something advanced, use OO

```python
heart=pd.read_csv("heart-disease.csv")
heart
```

```python
over_50 = heart[heart["age"] >50]
len(over_50)
```

```python
over_50.head()
```

## Pyplot plotting

```python
over_50.plot(kind="scatter",
            x="age",
            y="chol",
             c='target',
            figsize=(10,6));
```

## OO myplotlib plotting with pyplot

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot.scatter(x="age",
                     y="chol",
                     c="target",
                     ax=ax);
```

ax=ax -> telling to plat the this over_50 subplot in the ax upper plot which we created

## OO myplotlib

```python
over_50
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# PLot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
# custimize the plot
ax.set(title='Heart Disease',
      xlabel='Age',
      ylabel='Cholestrol');
# Add legend
ax.legend(*scatter.legend_elements(), title="target");

# Add a horizental line
ax.axhline(over_50['chol'].mean(),
          linestyle='--')
```

## Plotting Cholestrol & age, thalach & age

```python
heart
```

```python
#plotting subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(12,12))
#plotting data
ax1.scatter(heart["age"], heart["chol"], c=heart["target"])
#custimze ax1
ax1.set(title="Heart Diseases -> Cholestrol & Age",
       xlabel="Age",
       ylabel="Cholestrol")
#adding legend
ax1.legend(*scatter.legend_elements(), title="Target")
#adding mean value line
ax1.axhline(y=heart["chol"].mean(),
           linestyle="--")

#adding ax2 data
ax2.scatter(x=heart["age"], y=heart["thalach"], c=heart["target"])
#custimizing data
ax2.set(title="Max Heart Rate & Age",
       xlabel="Age",
       ylabel="Max Heart Rate");
#adding legend
ax2.legend(*scatter.legend_elements(), title="Target")
#adding mean line
ax2.axhline(y=heart["thalach"].mean(),
           linestyle="--")
# figure title
fig.suptitle("Heart Diseases Analysis", fontsize=16, fontweight="bold")
```
